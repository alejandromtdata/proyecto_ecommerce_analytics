```python
import pandas as pd
import numpy as np
from faker import Faker
import random

fake = Faker()

n = 100000
```

```python
paises_limpios = ["España", "Francia", "Italia", "Alemania"]

paises_sucios = {
    "España": ["España", "ES", "spain", "Espana"],
    "Francia": ["Francia", "FR", "france"],
    "Italia": ["Italia", "IT", "italy"],
    "Alemania": ["Alemania", "DE", "germany"]
}

clientes = []

for i in range(n):
    
    pais_base = random.choice(paises_limpios)
    pais = random.choice(paises_sucios[pais_base])
    
    if random.random() < 0.02:
        pais = "Unknown"

    nombre = fake.name()
    
    if random.random() < 0.05:
        nombre = None
    
    clientes.append({
        "id_cliente": i,
        "nombre": nombre,
        "pais": pais
    })

df_clientes = pd.DataFrame(clientes)

duplicados = df_clientes.sample(frac=0.03)
df_clientes = pd.concat([df_clientes, duplicados], ignore_index=True)

df_clientes.head()
```

```python
df_clientes.to_csv("../data/raw/clientes.csv", index=False)
```

```python
productos = ["PC", "Movil", "Tablet", "TV", "Auriculares"]

productos_sucios = {
    "PC": ["PC", "pc", "ordenador"],
    "Movil": ["Movil", "movil", "telefono"],
    "Tablet": ["Tablet", "tablet"],
    "TV": ["TV", "tv"],
    "Auriculares": ["Auriculares", "cascos"]
}

ventas = []

for i in range(n):

    id_cliente = random.randint(0, n-1)

    if random.random() < 0.02:
        id_cliente = 9999999

    prod_base = random.choice(productos)
    producto = random.choice(productos_sucios[prod_base])

    if random.random() < 0.02:
        producto = "unknown"

    fecha = fake.date_between(start_date='-2y', end_date='today')

    if random.random() < 0.03:
        fecha = str(fecha)

    importe = round(random.uniform(10, 2000), 2)

    if random.random() < 0.03:
        importe = str(importe)

    ventas.append({
        "id_venta": i,
        "fecha": fecha,
        "id_cliente": id_cliente,
        "producto": producto,
        "importe": importe
    })

df_ventas = pd.DataFrame(ventas)

df_ventas.head()
```

```python
df_ventas.to_csv("../data/raw/ventas.csv", index=False)
```

```python
productos_tabla = [
    {"producto": "PC", "categoria": "Dispositivos"},
    {"producto": "Movil", "categoria": "Dispositivos"},
    {"producto": "Tablet", "categoria": "Dispositivos"},
    {"producto": "TV", "categoria": "Dispositivos"},
    {"producto": "Auriculares", "categoria": "Audio"}
]

df_productos = pd.DataFrame(productos_tabla)

df_productos = pd.concat([df_productos, df_productos.iloc[[0]]], ignore_index=True)

df_productos = pd.concat([
    df_productos,
    pd.DataFrame([{"producto": "Consola", "categoria": "Gaming"}])
], ignore_index=True)

df_productos
```

```python
df_productos.to_csv("../data/raw/productos.csv", index=False)
```
